```markdown
# Подробный баланс для степенных корреляций

### Определения и выражения

Для доказательства того, что \( e_{kk'} = q_k P(k'|k) \), начнем с определения каждого термина через параметры сети:

1. **Совместная вероятность \( e_{kk'} \):**  
   Совместная вероятность \( e_{kk'} \) — это вероятность того, что случайно выбранное ребро соединяет два узла со степенями \( k \) и \( k' \):
   \[
   e_{kk'} = \frac{E_{kk'}}{L}
   \]
   где \( L \) — общее число рёбер в сети, а \( E_{kk'} \) — количество рёбер, соединяющих узлы степеней \( k \) и \( k' \). Если \( k = k' \), то \( E_{kk} \) равно удвоенному числу таких рёбер.

2. **Вероятность \( q_k \):**  
   \( q_k \) — вероятность того, что случайно выбранный конец ребра соединён с узлом степени \( k \):
   \[
   q_k = \frac{k N_k}{\langle k \rangle N}
   \]
   где \( N_k \) — число узлов со степенью \( k \), \( \langle k \rangle \) — средняя степень в сети, а \( N \) — общее число узлов.

3. **Условная вероятность \( P(k'|k) \):**  
   \( P(k'|k) \) — это условная вероятность того, что узел со степенью \( k \) соединён с узлом со степенью \( k' \):
   \[
   P(k'|k) = \frac{E_{kk'}}{k N_k}
   \]
   Это доля рёбер от узлов степени \( k \), которые соединяются с узлами степени \( k' \).

---

### Совмещение выражений

Мы хотим доказать, что:
\[
e_{kk'} = q_k P(k'|k)
\]

1. Подставим выражения для \( q_k \) и \( P(k'|k) \) в правую часть:
   \[
   q_k P(k'|k) = \left( \frac{k N_k}{\langle k \rangle N} \right) \left( \frac{E_{kk'}}{k N_k} \right)
   \]

2. Упростим выражение:
   \[
   q_k P(k'|k) = \frac{E_{kk'}}{\langle k \rangle N}
   \]

3. Учтем, что общее число рёбер \( L \) связано со средним степенным числом \( \langle k \rangle N \):
   \[
   L = \frac{\langle k \rangle N}{2}
   \]

4. Подставим \( L \) в выражение:
   \[
   q_k P(k'|k) = \frac{E_{kk'}}{L}
   \]

5. Заметим, что \( \frac{E_{kk'}}{L} \) — это \( e_{kk'} \):
   \[
   q_k P(k'|k) = e_{kk'}
   \]

---

### Заключение

Таким образом, мы доказали, что:
\[
e_{kk'} = q_k P(k'|k)
\]
```

```markdown
# Задача: Star Network

**Сеть "звезда"** состоит из одного центрального узла (хаба), который соединён с \( N - 1 \) периферийными узлами, каждый из которых имеет степень 1. Рассмотрим степень распределения, \( q_k \), и коэффициент корреляции степеней \( r \).

---

## 1. Распределение степеней \( p_k \)

- **Хаб** имеет степень \( N - 1 \).
- **Периферийные узлы** имеют степень 1.

Распределение степеней \( p_k \) выглядит следующим образом:
\[
p_k =
\begin{cases}
\frac{N - 1}{N}, & k = 1, \\
\frac{1}{N}, & k = N - 1, \\
0, & \text{иначе}.
\]

---

## 2. Вероятность \( q_k \)

Вероятность \( q_k \) — это вероятность того, что случайно выбранный конец ребра соединён с узлом степени \( k \). Она рассчитывается как:
\[
q_k = \frac{k N_k}{\langle k \rangle N},
\]
где:
- \( N_k \) — количество узлов со степенью \( k \),
- \( \langle k \rangle = 2 \) — средняя степень в сети,
- \( N \) — общее количество узлов.

Для сети "звезда":
- Для \( k = 1 \) (периферийные узлы):
  \[
  q_1 = \frac{1 \cdot (N - 1)}{2N} = \frac{N - 1}{2N}.
  \]
- Для \( k = N - 1 \) (хаб):
  \[
  q_{N-1} = \frac{(N - 1) \cdot 1}{2N} = \frac{N - 1}{2N}.
  \]

Обе вероятности равны, поскольку количество рёбер, идущих от хаба, совпадает с количеством рёбер, ведущих к периферийным узлам.

---

## 3. Коэффициент корреляции степеней \( r \)

Коэффициент корреляции степеней \( r \) определяется как:
\[
r = \frac{\sum_{k, k'} k k' \left( e_{kk'} - q_k q_{k'} \right)}{\sigma^2},
\]
где:
- \( e_{kk'} = \frac{E_{kk'}}{L} \) — совместная вероятность соединения узлов со степенями \( k \) и \( k' \),
- \( q_k q_{k'} \) — произведение независимых вероятностей нахождения степеней \( k \) и \( k' \),
- \( \sigma^2 = \sum_k k^2 q_k - \left( \sum_k k q_k \right)^2 \) — дисперсия степеней.

### Шаг 1: Совместная вероятность \( e_{kk'} \)
- Хаб (\( k = N-1 \)) соединён со всеми периферийными узлами (\( k = 1 \)). Таким образом:
  \[
  e_{1, N-1} = \frac{N - 1}{L},
  \]
  где \( L = N - 1 \) — общее количество рёбер в сети "звезда". Тогда:
  \[
  e_{1, N-1} = \frac{N - 1}{N - 1} = 1.
  \]
- Нет рёбер между узлами одинаковой степени, следовательно:
  \[
  e_{1,1} = 0, \quad e_{N-1, N-1} = 0.
  \]

### Шаг 2: Дисперсия \( \sigma^2 \)
- Рассчитаем среднее значение степени \( \mu_k = \sum_k k q_k \):
  \[
  \mu_k = 1 \cdot \frac{N - 1}{2N} + (N - 1) \cdot \frac{N - 1}{2N} = \frac{N - 1}{2} + \frac{(N - 1)^2}{2N}.
  \]
- Упростим выражение:
  \[
  \mu_k = \frac{N - 1}{2} \left( 1 + \frac{N - 1}{N} \right).
  \]

- Найдём дисперсию:
  \[
  \sigma^2 = \sum_k k^2 q_k - \mu_k^2.
  \]
  - Вклад от \( k = 1 \):
    \[
    k^2 q_k = 1^2 \cdot \frac{N - 1}{2N} = \frac{N - 1}{2N}.
    \]
  - Вклад от \( k = N - 1 \):
    \[
    k^2 q_k = (N - 1)^2 \cdot \frac{N - 1}{2N}.
    \]
  - Итоговая сумма:
    \[
    \sum_k k^2 q_k = \frac{N - 1}{2N} + \frac{(N - 1)^3}{2N}.
    \]

### Шаг 3: Корреляция \( r \)
Так как в сети "звезда" все рёбра соединяют хаб с периферийными узлами, а рёбра между узлами одинаковой степени отсутствуют, корреляция оказывается отрицательной:
\[
r < 0,
\]
что указывает на **дизассоциативность**.

---

## 4. Ассоциативна или дизассоциативна?

Сеть "звезда" является **дизассоциативной**, так как:
- Узлы высокой степени (хаб) соединяются только с узлами низкой степени (периферийными).
- Связи между узлами одинаковой степени отсутствуют, что усиливает дизассоциативный характер.

---

## Итог

1. **Распределение степеней \( p_k \):**
   \[
   p_k =
   \begin{cases}
   \frac{N-1}{N}, & k = 1, \\
   \frac{1}{N}, & k = N-1, \\
   0, & \text{иначе}.
   \]

2. **Вероятности \( q_k \):**
   - Для \( k = 1 \): \( q_1 = \frac{N - 1}{2N} \),
   - Для \( k = N-1 \): \( q_{N-1} = \frac{N - 1}{2N} \).

3. **Коэффициент корреляции \( r \):**
   \[
   r < 0 \ (\text{дизассоциативность}).
   \]

4. **Вывод:** Сеть "звезда" является дизассоциативной, поскольку хаб соединён только с периферийными узлами, и отсутствуют связи между узлами одинаковой степени.
```

# Степенные корреляции в сетях Эрёди-Реньи

### 1. Модель \( G(N, L) \)

#### Вероятность связи \( e_{ij} \)
В модели \( G(N, L) \):
- \( N \) узлов соединены \( L \) случайно распределёнными рёбрами.
- Вероятность того, что два узла \( (i, j) \) связаны, равна:
  \[
  e_{ij} = \frac{L}{{N \choose 2}} = \frac{2L}{N(N-1)}.
  \]

#### Условная вероятность \( P(i, j \,|\, l, s) \)
Условная вероятность \( P(i, j \,|\, l, s) \) зависит от того, являются ли связи \( (i, j) \) и \( (l, s) \) различными:
1. Если \( (i, j) \neq (l, s) \), существование связи \( (l, s) \) уменьшает количество оставшихся рёбер для размещения \( L - 1 \):
   \[
   P(i, j \,|\, l, s) = \frac{L-1}{{N \choose 2} - 1}.
   \]

2. Если \( (i, j) = (l, s) \), то:
   \[
   P(i, j \,|\, l, s) = 1,
   \]
   так как эта связь предопределена.

---

### 2. Отношение вероятностей

Отношение двух вероятностей для \( (i, j) \neq (l, s) \):
\[
\text{Отношение} = \frac{P(i, j \,|\, l, s)}{e_{ij}} = \frac{\frac{L-1}{{N \choose 2} - 1}}{\frac{L}{{N \choose 2}}}.
\]

Упростим:
\[
\text{Отношение} = \frac{(L-1)(N(N-1))}{L(N(N-1) - 2)} = \frac{L-1}{L} \cdot \frac{N(N-1)}{N(N-1) - 2}.
\]

#### Поведение для малых и больших сетей

1. **Малые сети (\( N \to 2 \)):**
   - \( N(N-1) - 2 \to 0 \), поэтому знаменатель становится намного меньше числителя.
   - Отношение сильно возрастает, что указывает на более сильную зависимость между рёбрами.

2. **Большие сети (\( N \to \infty \)):**
   - \( N(N-1) \approx N(N-1) - 2 \), поэтому:
     \[
     \text{Отношение} \approx \frac{L-1}{L} \to 1.
     \]
   - Зависимость между рёбрами становится незначительной.

---

### 3. Модель \( G(N, p) \)

В модели \( G(N, p) \):
- Каждая пара узлов \( (i, j) \) соединена с вероятностью \( p \), независимой от других связей.

#### Вероятность связи \( e_{ij} \)
\[
e_{ij} = p.
\]

#### Условная вероятность \( P(i, j \,|\, l, s) \)
Так как рёбра размещаются независимо:
\[
P(i, j \,|\, l, s) = p.
\]

#### Отношение вероятностей
\[
\text{Отношение} = \frac{P(i, j \,|\, l, s)}{e_{ij}} = \frac{p}{p} = 1.
\]

Для модели \( G(N, p) \) отсутствует зависимость между рёбрами независимо от размера сети.

---

### 4. Сравнение моделей \( G(N, L) \) и \( G(N, p) \)

#### Зависимость между рёбрами
- В модели \( G(N, L) \):
  - Рёбра размещаются не независимо, так как общее число рёбер фиксировано (\( L \)).
  - Для малых сетей размещение одного ребра существенно влияет на размещение других.
  - Для больших сетей зависимость становится незначительной при увеличении \( N \).

- В модели \( G(N, p) \):
  - Рёбра размещаются независимо, что исключает любую зависимость между связями.

#### Последствия для малых сетей
- Для малых сетей (\( N \to 2 \)):
  - Модель \( G(N, L) \) вводит сильные корреляции между рёбрами, что может привести к отклонению свойств сети от случайных.
  - Модель \( G(N, p) \) обеспечивает истинную случайность, что делает её предпочтительной для генерации малых случайных сетей.

#### Последствия для больших сетей
- Для больших сетей (\( N \to \infty \)):
  - Обе модели дают схожие результаты, так как зависимость между рёбрами в \( G(N, L) \) становится незначительной.

---

### Итог

- **Малые сети:** Модель \( G(N, p) \) предпочтительна для генерации случайных сетей из-за независимости размещения рёбер.
- **Большие сети:** Обе модели \( G(N, L) \) и \( G(N, p) \) подходят, так как различия в зависимости между рёбрами становятся несущественными.
